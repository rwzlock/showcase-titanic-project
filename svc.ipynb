```python
train_data = train_data.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
test_data = test_data.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
train_data.shape
```

```python
X = train_data[["Sex","Pclass","SibSp", "Age", "Fare"]]
y = train_data["Survived"].values.reshape(-1, 1)
target_names = ["Survived", "Not Survived"]
print(X.shape, y.shape)
```

```python
X = pd.get_dummies(X, columns=["Sex"])
X.head()
```

```python
# Create the bins in which Data will be held
# Bins are 0 to 25, 25 to 50, 50 to 75, 75 to 100
bins = [0, 100, 200, 300, 400, 550]

# Create the names for the four bins
fare_group_names = ['Vey Low',"Low", 'Okay', 'High', 'Highest']
```

```python
X["Fare"] = pd.cut(X["Fare"], bins, labels=fare_group_names)
X.head()
```

```python
X = pd.get_dummies(X, columns=["Fare"])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, random_state=1)

X_train, X_test, y_train, y_test = train_test_split(test, train, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=7))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
deep_model = Sequential()
deep_model.add(Dense(units=6, activation='relu', input_dim=7))
deep_model.add(Dense(units=6, activation='relu'))
deep_model.add(Dense(units=6, activation='relu'))

deep_model.add(Dense(units=2, activation='softmax'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
